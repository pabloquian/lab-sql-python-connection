## Challenge
**In this lab, the objective is to identify the customers who were active in both May and June of 2005, and how did their activity differ between months.** To achieve this, follow these steps:

1. Establish a connection between Python and the Sakila database.

2. Write a Python function called rentals_month that retrieves rental_date data for a given month and year (passed as parameters) from the Sakila database as a Pandas DataFrame. The function should take in three parameters:

    - engine: an object representing the database connection engine to be used to establish a connection to the Sakila database.
    - month: an integer representing the month for which rental data is to be retrieved.
    - year: an integer representing the year for which rental data is to be retrieved.
The function should execute a SQL query to retrieve the rental data for the specified month and year from the rental table in the Sakila database, and return it as a pandas DataFrame.

3. Develop a Python function called rental_count_month that takes the DataFrame provided by rentals_month as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id during the selected month and year.

The function should also include the month and year as parameters and use them to name the new column according to the month and year, for example, if the input month is 05 and the year is 2005, the column name should be "rentals_05_2005".


4. Create a Python function called compare_rentals that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

```python
# First I import the libraries I will need
import pandas as pd
import numpy as np

import pymysql                        # for getting data from a SQL database
from sqlalchemy import create_engine  # for establishing the connection and authentication

from getpass import getpass  # To get the password without showing the input    
```

```python
# Let's see if I have access to sakila database
data = pd.read_sql_query('SELECT * FROM sakila.film', engine)
data.head()
```

```python
# Now I will create a function called rentals_month that retrieves rental_date data for a given month and year
# (passed as parameters) from the Sakila database as a Pandas DataFrame.
def rentals_month (year, month, engine_db):
        
        
        if len(str(month)) == 1:
            month_str = "0" + str(month)
        else:
            month_str = str(month) 
        
        query = '''SELECT * 
                FROM sakila.rental 
                WHERE DATE_FORMAT(rental_date, "%%Y-%%m") = CONCAT("'''+str(year)+'","-", "'+month_str+'")'
        
        result = engine.execute(query)
        result_df = pd.DataFrame(result)
        return result_df
    
```

```python
# Let's try if the function works
rentals_month(2005,5,engine)
```

```python
# I will define a function that makes the month a str with a 0 at the beggining if it's a month between 1 and 9.

def format_month_str(month):
    '''
    This function converts a month to a string, adding a 0 at the beginning to months 1 to 9.
    Input:
        month: a number that represents a month
    Output:
        that number converted to a string and with a 0 at the beginning if the month is between 1 (January) and 9 (September)
    '''
    if len(str(month)) == 1:
        return "0" + str(month)
    else:
        return str(month)
```

```python
# Next step in the lab is develop a function called rental_count_month that takes the DataFrame provided by rentals_month  
# as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id 
# during the selected month and year.

def rental_count_month (year, month, engine_db):
    
    df_rentals = rentals_month(year, month, engine_db) 
    
    # Count the number of rentals made for each customer_id
    rentals_customer = df_rentals['customer_id'].value_counts() 
    
    # Convert to Dataframe
    rentals_customer_df = pd.DataFrame(rentals_customer)
    
    # Reset the index
    rentals_customer_df.reset_index(inplace=True)
     
    # Rename the columns
    rentals_customer_df.rename(columns={'count': 'rentals_'+format_month_str(month)+'_'+str(year)}, inplace=True)
    
    return rentals_customer_df
```

```python
# I assing a new variable that contains a dataframe with the total rentals made by each customer in May of 2005
df_2005_05 = rental_count_month(2005, 5, engine)

# And I assing another variable that contains the same info corresponding to June of 2005.
df_2005_06 = rental_count_month(2005, 6, engine)
```

```python
# I will build a function called compare_rentals that takes two DataFrames as input containing the number of rentals made 
# by each customer in different months and years.

def compare_rentals (df_1, df_2):
    
    # First I merge both dataframes on customer_id. I add how='outer' to include the customers that are present in one dataframe but missing in the other.
    merged_df = pd.merge(df_1, df_2, on='customer_id', how='outer')
    
    # Fill NA with 0 if any, in order to perform correctly the substraction correctly
    merged_df.iloc[:, 1] = merged_df.iloc[:, 1].fillna(0)
    merged_df.iloc[:, 2] = merged_df.iloc[:, 2].fillna(0)
    
    # Add a new column that is the total rentals made by a customer in a certain month minus the total rentals made in another month.
    merged_df['difference'] = merged_df.iloc[:, 1] - merged_df.iloc[:, 2]
    
    return merged_df
```

```python
# Now let's see the difference of rentals made by customers in May of 2005 and June of 2005.
diff_may_jun_2005 = compare_rentals(df_2005_05, df_2005_06)
diff_may_jun_2005
```
